```python
lst = df.crypto_name.unique()
```

```python
def process(lst, dt):
    for i in lst:
        if i == "Bitcoin":
            data = df.loc[df['crypto_name']==i]
            data = data.iloc[[-1]]
            dt = data
        else:
            data = df.loc[df['crypto_name']==i]
            dt = pd.concat([dt,data.iloc[[-1]]])
    return dt
```

```python
def extract(df):
    lst = df.crypto_name.unique()
    dt = pd.DataFrame()
    return process(lst, dt)
```

```python
dt = extract(df)
```

```python
new_dt = dt.sort_values(by='close', ascending = False)
```

```python
new_dt = new_dt.set_index(new_dt['crypto_name'])
```
